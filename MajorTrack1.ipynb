```python
import os
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Import required libraries
from roboflow import Roboflow
from ultralytics import YOLO

# 1. DOWNLOAD THE DATASET
print("Downloading dataset from Roboflow...")
rf = Roboflow(api_key="lEsJ7Y0zN3X4BV6U54L6")
project = rf.workspace("meryem-ece-ayyildiz-p8ji1").project("od-of-aquarium-fishes")
version = project.version(1)
dataset = version.download("yolov8")

print(f"Dataset downloaded to: {dataset.location}")

# --- SAFE PATH FINDER (Windows Uyumlu Dosya Bulucu) ---
yaml_path = os.path.join(dataset.location, "data.yaml")

# Eğer dosya ana dizinde yoksa, alt klasörlerin içine bakıp bulur
if not os.path.exists(yaml_path):
    yaml_files = glob.glob(os.path.join(dataset.location, "**", "data.yaml"), recursive=True)
    if len(yaml_files) > 0:
        yaml_path = yaml_files[0]
    else:
        raise FileNotFoundError("\nHATA: data.yaml dosyasi inmemis! Lutfen klasoru silip kodu tekrar calistirin.")

print(f"Safe YAML path established at: {yaml_path}")
# ------------------------------------------------------

# 2. LOAD THE PRE-TRAINED MODEL
print("Loading YOLOv8n model...")
model = YOLO('yolov8n.pt') 

# 3. TRAIN THE MODEL ON YOUR CUSTOM DATASET
# (This process may take 15-30 minutes on a CPU. Grab a coffee and wait!)
print("Training YOLOv8 model...")
results = model.train(data=yaml_path,  # <--- GÜNCELLENDİ (Otomatik bulunan yol)
                      epochs=10, 
                      imgsz=416, 
                      plots=True) 

# 4. EVALUATE MODEL PERFORMANCE
print("\nEvaluating model performance...")
metrics = model.val()
print(f"mAP50-95 (Overall Mean Average Precision): {metrics.box.map:.4f}")
print(f"mAP50 (Mean Average Precision at IoU 0.5): {metrics.box.map50:.4f}")

# 5. RUN INFERENCE ON TEST IMAGES
print("\nRunning inference on test images...")
test_images_path = os.path.join(dataset.location, "test", "images")
prediction_results = model.predict(source=test_images_path, save=True, conf=0.25)

# 6. DISPLAY SAMPLE PREDICTIONS
print("\nDisplaying sample predictions:")
predict_folders = sorted(glob.glob("runs/detect/predict*"), key=os.path.getmtime)
latest_predict_folder = predict_folders[-1]

sample_images = glob.glob(os.path.join(latest_predict_folder, "*.jpg"))[:3]

plt.figure(figsize=(15, 10))
for i, img_path in enumerate(sample_images):
    img = mpimg.imread(img_path)
    plt.subplot(1, 3, i+1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Test Image {i+1}")
plt.tight_layout()
plt.show()
```
